```python
# 🚀 Import libraries
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# 🧠 Load the Mistral-7B-Instruct model (8-bit for less RAM usage)
model_name = "mistralai/Mistral-7B-Instruct-v0.3"
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load model in 8-bit to avoid RAM crashes
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    load_in_8bit=True,               # Load in 8-bit to save memory
    device_map="auto"               # Use GPU if available
)

# 🟢 Create a text-generation pipeline
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

# ✍️ Test the chatbot with a sample question
prompt = "What are the main steps in data preprocessing?"
response = pipe(prompt, max_length=200, do_sample=True, temperature=0.7)

# 🖨️ Print the response
print(response[0]['generated_text'])
```
